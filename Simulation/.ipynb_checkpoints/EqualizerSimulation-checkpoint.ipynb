# FPGA Equalizer Simulation

For testing and aiding in the design of the DSP portions of the FPGA Equalizer 

```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
```

```python
# user defined parameters
filter_count = 4

# variables
filter_sizes = np.zeros(filter_count, dtype=int)
fixed_point_precisions = np.zeros(filter_count, dtype=int)
filter_max_values = np.zeros(filter_count, dtype=int)
filter_taps = np.empty((0, int(filter_sizes[0])), dtype=int)
```

```python
# read filter taps from text file
for index in range(filter_count) :
    file_name = "filter{}_taps.txt".format(index+1)
    file = open(file_name, "r")
    text = file.readlines()
    filter_sizes[index] = int(text[0][1:])
    fixed_point_precisions[index] = int(text[1][1:])
    taps = np.array([int(element) for element in text[2:]])
    taps_size_check = len(taps)
    if (taps_size_check != filter_sizes[index]) :
        print("Error: sizes don't match up!")
        print("Defined filter size:", filter_sizes[index])
        print("Actual filter size:", taps_size_check)
    center_tap_index = int(filter_sizes[index] / 2)
    filter_max_values[index] = taps[center_tap_index]
    filter_taps = np.array([filter_taps, taps], dtype=object)
    file.close()
    
    # print results
    print("Filter{} Size:".format(index+1), filter_sizes[index], "taps")
    print("Filter{} Fixed-Point Precision:".format(index+1), fixed_point_precisions[index], "bits")
    print("Filter{} Maximum Tap Value:".format(index+1), filter_max_values[index])
    print("")
```

```python
for taps_iter in filter_taps:
    taps = []
    if (taps_iter == filter_taps[0]) :
        taps = taps_iter[1]
    else:
        taps = taps_iter

    # get frequency response
    w, H = signal.freqz(taps)
    angles = np.unwrap(np.angle(H))

    # plot frequency response
    fig = plt.figure()
    plt.title('Filter Magnitude Response')
    plt.plot(w, 20 * np.log10(abs(H)))
    plt.ylabel('Amplitude [dB]')
    plt.xlabel('Frequency [rad/sample]')
    plt.show()

    fig = plt.figure()
    plt.title('Filter Phase Response')
    plt.plot(w, angles)
    plt.ylabel('Angle (radians)')
    plt.xlabel('Frequency [rad/sample]')
    plt.show()
```
